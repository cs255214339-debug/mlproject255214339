```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import VarianceThreshold
# Step 1: Create dataset with missing values
data = {
    "name": ["Alice", "Bob", "Charlie", "David", "Eva",
             "Frank", "Grace", "Hank", "Ivy", "Jack"],
    "age": [25, 30, 35, 40, 28, 33, 29, 45, 31, 38],
    "height": [165, 175, 180, 170, 160, 178, np.nan, 172, 168, 182],
    "bp": [120, 130, np.nan, 140, 110, 135, 125, 150, 118, 145],
    "weight": [55, 70, 80, 75, 50, 78, 60, 85, 58, 82]
}
df = pd.DataFrame(data)
# Step 2: Handle missing values (mean imputation for numerical columns)
df['height'].fillna(df['height'].mean(), inplace=True)
df['bp'].fillna(df['bp'].mean(), inplace=True)
# Step 3: Normalize numerical columns using Min-Max scaling
scaler = MinMaxScaler()
numeric_cols = ['age', 'height', 'bp', 'weight']
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

# Step 4: Feature selection (remove low-variance features)
selector = VarianceThreshold(threshold=0.01)
selected_features = selector.fit_transform(df[numeric_cols])
selected_cols = [col for col, keep in zip(numeric_cols, selector.get_support()) if keep]
# Final preprocessed dataset
preprocessed_df = pd.concat([df[['name']], df[selected_cols]], axis=1)
print("✅ Preprocessed Dataset:")
print(preprocessed_df)
```
